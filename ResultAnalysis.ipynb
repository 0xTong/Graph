```python
import pandas as pd
import numpy as np

# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'iso')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['iso']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/7.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("iso")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print(" pearson")
print(df.corr("pearson"))

#M3
# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'M3')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['M3']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/1.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("\nM3")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print("pearson")
print(df.corr("pearson"))
#M2
# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'M2')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['M2']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/5.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("\nM2")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print("pearson")
print(df.corr("pearson"))

#M1
# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'M1')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['M1']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/9.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("\nM1")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print("pearson")
print(df.corr("pearson"))


#R1
# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'R1')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['R1']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/14.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("\nR1")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print("pearson")
print(df.corr("pearson"))

#R2
# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'R2')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['R2']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/16.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("\nR2")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print("pearson")
print(df.corr("pearson"))

#R3
# ground truth
df_gt = pd.read_excel("./test/Answer.xlsx", 'R3')

wavelength_iso_gt = df_gt['nm']
value_iso_gt = df_gt['R3']

# prediction_iso
df_pred_iso = pd.read_csv("./test/DataResult/12.csv")
wavelength_iso_pred = np.array(df_pred_iso['x'])
value_iso_pred = np.array(df_pred_iso['y'])

print("\nR3")
total_err = 0
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        total_err += abs(value_iso_gt[i] - value_iso_pred[np.argwhere(wavelength_iso_pred == wavelength)])**2

mse = total_err / len(wavelength_iso_gt)
print("MSE: ", mse)
print()

df_corr =[]
df_corr_pred =[]
for i in range(len(wavelength_iso_gt)):
    wavelength = wavelength_iso_gt[i]
    if wavelength in wavelength_iso_pred:
        df_corr.append(value_iso_gt[i])
 

i=0
wavelenth = 0

for i in range(len(wavelength_iso_pred)):
    wavelength = wavelength_iso_pred[i]
    if wavelength in np.array(wavelength_iso_gt):
        df_corr_pred.append(df_pred_iso['y'][i])
list_tuples = list(zip(df_corr, df_corr_pred))  
df = pd.DataFrame(list_tuples)
print("pearson")
print(df.corr("pearson"))
```
